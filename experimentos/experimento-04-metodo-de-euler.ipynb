# Metodo de Euler

Uma outra forma de calcular EDO de primeira ordem para problemas de valores iniciais.

O nosso problema:



Para x ∈ [0,1] e N = 4

---

Revisão método de Euler:




```python
def edo(x, y):
  return y - x
```

```python
x0 = 0
y0 = 2
```

```python
I = [0,1]
```

```python
N = 4
```

```python
h = (I[1] - I[0]) / N
```

```python
xi = [x0]
yi = [y0]
```

```python
for index in range(N):
  yk = y0 + h * edo(x0, y0)
  y0 = yk
  x0 = x0 + h

  xi.append(x0)
  yi.append(y0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(xi, yi)
```

```python
plt.plot(xi, yi, 'o')
```
