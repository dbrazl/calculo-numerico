# Integral de uma função

Será usado o método dos trapézios.

$\int_a^b f(x) \, dx \approx \frac{h}{2} \left( f(x_0) + 2 \sum_{i=1}^{n-1} f(x_i) + f(x_n) \right)$

$\equiv$

$\int_a^b f(x) \, dx \approx \frac{h}{2} \left( f(x_0) + f(x_n) \right) + \frac{2h}{2} \left(\sum_{i=1}^{n-1} f(x_i) \right)$

$\equiv$

$\int_a^b f(x) \, dx \approx \frac{h}{2} \left( f(x_0) + f(x_n) \right) + h \left(\sum_{i=1}^{n-1} f(x_i) \right)$

```python
import numpy as np
```

```python
def integrate(f, a, b, h):
  n = (b - a) / h
  x = np.linspace(a, b, int(n) + 1)

  return 0.5 * h * (f(x[0]) + f(x[-1])) + h * np.sum(f(x[1:-1]))
```

Exemplo 1: $\int_2^4 x^2 \text{dx}$

```python
def f(x):
  return x**2
```

```python
integrate(f, 2, 4, 0.01)
```

Exemplo 2: $\int_0^{10} \sin{x} dx$

```python
def f(x):
  return np.sin(x)
```

```python
integrate(f, 0, 10, 0.01)
```
